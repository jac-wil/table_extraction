# demo.ipynb

Demonstration of synthetic OCR generation, slope correction, and table reconstruction.

```python
# ────────────────────────────────
# 1. Setup & Imports
# ────────────────────────────────
from generate_synthetic import generate_synthetic
from table_utilities import OCRBox, load_ocr_json, correct_slope
from table_reconstruction import plot_table_grid, plot_table_grid_auto
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# ────────────────────────────────
# 2. Generate Synthetic OCR Data
# ────────────────────────────────
generate_synthetic(out_path="sample_data/ocr_output.json")

boxes = load_ocr_json("sample_data/ocr_output.json")
```

```python
# ────────────────────────────────
# 3. Detect Headers/Table Detail/Footers & Visualize Before/After Slope Correction
# ────────────────────────────────
corrected = correct_slope(boxes)

headers = ["Units", "Pieces", "Weight", "Description", "Total"]
footers = ["Total Charge", "Total Chg."]

print("Quadrant-Based Skewed Table Reconstruction:")
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
plot_table_grid(ax1, boxes, headers, footers, "Skewed Table")
plot_table_grid_auto(ax2, corrected, headers, footers, "Reconstructed Tabular Data")
plt.tight_layout()
plt.show()
```
